the computer responds with its move and the check is repeated;
don't implement any form of artificial intelligence − a random field choice made by the computer is good enough for the game.

Requirements

Implement the following features:

* the board should be stored as a three-element list, while each element is another three-element list (the inner lists represent rows) so that all of the squares may be accessed using the following syntax:

```python
board[row][column]
```

* each of the inner list's elements can contain 'O', 'X', or a digit representing the square's number (such a square is considered free)
* the board's appearance should be exactly the same as the one presented in the example.
* implement the functions defined for you in the editor.

Drawing a random integer number can be done by utilizing a Python function called randrange(). The example program below shows how to use it (the program prints ten random numbers from 0 to 8).

Note: the from-import instruction provides access to the randrange function defined within an external Python module callled random.

```python
from random import randrange
```

```python
from random import randrange

board = [[1,2,3],[4,"X",6],[7,8,9]]

def display_board(board):
  
  linha = "+-------+-------+-------+"
  separador = "|       |       |       |"
  separador_numero_1 = "|   "+str(board[0][0])+"   |   "+str(board[0][1])+"   |   "+str(board[0][2])+"   |"
  separador_numero_2 = "|   "+str(board[1][0])+"   |   "+str(board[1][1])+"   |   "+str(board[1][2])+"   |"
  separador_numero_3 = "|   "+str(board[2][0])+"   |   "+str(board[2][1])+"   |   "+str(board[2][2])+"   |"
  print(f"{linha}\n{separador}\n{separador_numero_1}\n{separador}\n{linha}")
  print(f"{separador}\n{separador_numero_2}\n{separador}\n{linha}")
  print(f"{separador}\n{separador_numero_3}\n{separador}\n{linha}")

def enter_move(board):
  jogo = make_list_of_free_fields(board)
  while True:
    try:
        print("Digite a posição da sua jogada de 1 a 9.")
        jogada = int(input("Faça a sua jogada: "))
        if jogada < 10 and jogada > 0:
          if jogada in jogo:
            break
          else:
            print("Essa posição já foi escolhida.")
    except:
        print("Digite uma jogada válida.")
  for i in range(3):
    for n in range(3):
      if jogada == board[i][n]:
        board[i][n] = "O"

def make_list_of_free_fields(board):
  jogo = []
  for i in range(3):
    for n in range(3):
      jogo.append(board[i][n])
  return jogo

def draw_move(board):
  jogo = make_list_of_free_fields(board)
  while True:
    computador = randrange(10)
    if computador in jogo:
      for i in range(3):
        for n in range(3):
          if computador == board[i][n]:
            board[i][n] = "X"
      break

def victory_for(board, sign):
  jogo = make_list_of_free_fields(board)
  jogador = jogo.count("O")
  computador = jogo.count("X")
  if (jogador + computador) == 9:
    return False
  else:
    jogadas = []
    vitoria = [[0,1,2],[3,4,5],[6,7,8],[0,4,8],[2,4,6],[0,3,6],[1,4,7],[2,5,8]]
    for i, n in enumerate(jogo):
      if n == sign:
        jogadas.append(i)
    for i in vitoria:
      if jogadas == i:
        return True
        break
    
```

```python
display_board(board)

while True:
  enter_move(board)
  display_board(board)
  #print(victory_for(board, "O"))
  if victory_for(board, "O") == True:
    print("O Jogador Venceu.")
    break
  elif victory_for(board, "O") == False:
    print("Houve Empate")
    break
  draw_move(board)
  display_board(board)
  #print(victory_for(board, "X"))
  if victory_for(board, "X") == True:
    print("O Computador Venceu.")
    break
  elif victory_for(board, "X") == False:
    print("Houve Empate")
    break
```
